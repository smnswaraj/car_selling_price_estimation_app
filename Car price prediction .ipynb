```python
# Droping un-useful features: 

final_dataset = df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven','Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
final_dataset.head()
```

```python
# Adding coloumn for showing current-year :

final_dataset['current_year'] = 2020
```

```python
# Adding coloumn to show how old is the car :

final_dataset['no_of_years'] = final_dataset['current_year'] - final_dataset['Year']
```

```python
final_dataset
```

```python
# Droping un-wanted coloumns :

final_dataset.drop(['Year','current_year'], axis=1, inplace=True)
```

```python
final_dataset
```

```python
# Converting categorical Fuel-Type column into dummy or indicator variables: 

final_dataset = pd.get_dummies(final_dataset)
```

```python
final_dataset
```

```python
# Selecting data for training and testing purpose in 80:20 rario:

X = final_dataset.iloc[:,1:]
Y = final_dataset.iloc[:,1]

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
# Trainig the Random Forest model with the help of training data :

from sklearn.ensemble import RandomForestRegressor

model= RandomForestRegressor()
model.fit(X_train,Y_train)
```

```python
# Checking Accuracy :
from sklearn.metrics import mean_squared_error

X_test_prediction = model.predict(X_test)
mean_squared_error(X_test_prediction,Y_test)
```

```python
# Importing the model as pickle file for deployement purpose:
import pickle 
with open('car_price_prediction_model.pkl','wb') as f: 
    pickle.dump(model,f)
```
